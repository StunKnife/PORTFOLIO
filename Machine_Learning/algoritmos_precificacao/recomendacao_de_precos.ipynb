## Variáveis
    - Como podemos ver, este conjunto de dados contém várias colunas:
        1. marca: marca de um carro (BMW, Toyota, e assim por diante)
        2. modelo: modelo de um carro
        3. ano: ano em que o carro foi fabricado
        4. engine_fuel_type: tipo de combustível que o motor necessita (diesel, elétrico, etc.)
        5. motor_hp: potência do motor
        6. engine_cylinders: número de cilindros no motor
        7. Transmission_type: tipo de transmissão (automática ou manual)
        8. rodas_acionadas: dianteira, traseira, todas
        9. number_of_doors: número de portas que um carro possui
        10. market_category: luxo, crossover e assim por diante
        11. tamanho_do_veículo: compacto, médio ou grande
        12. estilo_veículo: sedã ou conversível
        13. Highway_mpg: milhas por galão (mpg) na rodovia
        14. city_mpg: milhas por galão na cidade
        15. popularidade: número de vezes que o carro foi mencionado em um stream do Twitter
        16. msrp: preço de varejo sugerido pelo fabricante

    . Para nós, a coluna mais interessante aqui é a última: MSRP (preço de varejo sugerido pelo fabricante, ou simplesmente preço de um carro). Usaremos esta coluna para prever os preços de um carro

## Estrutura de Validação
    - Precisamos dividir o conjunto de dados em três partes: treinamento, validação e teste

```python
# Total de registros
n = len(df) 

# Tamanhos dos conjuntos
n_val = int(0.2 * n) 
n_test = int(0.2 * n) 
n_train = n - (n_val + n_test) 

# Fixando semente
np.random.seed(2) 

# Array de índeces
idx = np.arange(n) 

# Misturando os índeces
np.random.shuffle(idx) 

# Locanlizando os dados com base nos índeces
df_shuffled = df.iloc[idx] 

# Misturando os índices e dividindo os dados em três parcelas: 0 : x; x: y; y:z
df_train = df_shuffled.iloc[:n_train].copy() 
df_val = df_shuffled.iloc[n_train:n_train+n_val].copy() 
df_test = df_shuffled.iloc[n_train+n_val:].copy()
```

```python
# Transformação logarítmica para cada dataframe
y_train = np.log1p(df_train.msrp.values)
y_val = np.log1p(df_val.msrp.values)
y_test = np.log1p(df_test.msrp.values)
```

```python
# Para evitar o uso acidental da target posteriormente, vamos removê-la dos dataframes
del df_train['msrp']
del df_val['msrp']
del df_test['msrp']
```

## Análise de Correlação

```python
# Análise de correlação
numeric_columns = df.select_dtypes(include=['int64', 'float64']).drop("number_of_doors",axis=1)
numeric_columns
```

```python
# Calcula a matriz de correlação
correlation_matrix = df.drop("number_of_doors",axis=1).corr()
msrp_correlations = correlation_matrix['msrp'].abs().sort_values(ascending=False) # Correlações com a variável resposta 
msrp_correlations
```

```python
# Filtrando as colunas com maiores correlações
# Define um limiar de correlação
threshold = 0.15

# Filtra as colunas com valores absolutos de correlação menores ou iguais ao limiar - objetivo dropar
significant_correlations = msrp_correlations[msrp_correlations.abs() <= threshold]

# Armazena as colunas em um objeto 
selected_num_columns = significant_correlations.index

# Exibe as colunas selecionadas
print(selected_num_columns)
```

## Variáveis Categóricas

```python
# Selecionando as colunas categóricas
object_columns = df_train.select_dtypes(include=['object']).columns.tolist()

# Adicionando número de portas no grupo de categóricas
object_columns.append('number_of_doors')

# Obtendo as dummies
df_train = pd.get_dummies(df_train, columns=object_columns,drop_first=True)#.drop('year',axis=1)

df_train
```

# Regressão Linear Múltipla
    - Para prever o preço de um carro, precisamos usar um modelo de aprendizado de máquina. Para fazer isso, usaremos a regressão linear.

```python
selected_num_columns
```

```python
# Adicionando uma constante de intercepto ao modelo - coluna de valores 1
X = sm.add_constant(df_train)

# Dropando algumas variáveis numéricas com correlação baixa
X = X.drop(columns=selected_num_columns,axis=1)
X
```

```python
X = X.drop(columns='engine_fuel_type_electric',axis=1)
```

```python
# Ajuste o modelo de regressão
model = sm.OLS(y_train, X).fit()

# Obtenha o resumo da regressão
summary = model.summary()

# Exiba o resumo
print(summary) # Popularity e highway_mpg não foram significativos                                                   

```

```python
import statsmodels.api as sm
import pandas as pd

# Ajuste o modelo de regressão
model = sm.OLS(y_train, X).fit()

# Obtenha os resultados da regressão
results = model.summary()

# Obtenha os valores de p-valor
p_values = model.pvalues

# Crie um DataFrame com o nome das variáveis e os valores de p-valor
p_values_df = pd.DataFrame({'Variable': p_values.index, 'P-Value': p_values.values}).sort_values(by='P-Value', ascending=False)
p_values_df = p_values_df[p_values_df['P-Value']>=0.05]
# Exiba o DataFrame
print(p_values_df)
```
